```python
import matplotlib.pyplot as plt
```

```python
x = [1, 3]
```

```python
y = [2, 2]
```

```python
plt.figure(figsize=(6, 4))
plt.scatter(x, y, s=800, color='gold',
            edgecolor='pink', linewidths=2, alpha=0.9)
sparkle_x = [1.1, 2.9]
sparkle_y = [2.2, 2.2]
plt.scatter(sparkle_x, sparkle_y, s=100,
            color='white', edgecolor='gold',
            linewidths=1.5)
plt.title(" First Sight - Her Eyes in the West Corridor ", fontsize=16,
                                          fontweight='bold')
plt.text(1.2, 1.5, "Golden Calm\nthrough the corridor light", fontsize=12,
         color='darkgoldenrod')
plt.xlim(0, 4)
plt.ylim(1, 3)
plt.axis('off')

plt.show()
```
